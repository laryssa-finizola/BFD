```python
Slicing (ou fatiamento) em Python é uma técnica que permite extrair uma parte (subsequência) de uma sequência, como uma string ou lista
```

SINTAXE: [início:fim:passo]

```python
#EXEMPLO: 

s = "Python"

print(s[1:4])   # 'yth' pega índice 1 até 3
print(s[:3])    # 'Pyt' do começo até índice 2
print(s[3:])    # 'hon' do índice 3 até o fim
print(s[::2])   # 'Pto' pula 1 em 1 caractere
print(s[::-1])  # 'nohtyP' string invertida
```
